#### Total number of reported cases and deaths based on the data:-

```python
total_cases=sum(covid_df.new_cases)
total_deaths=sum(covid_df.new_deaths)
```

```python
total_cases
```

```python
total_deaths
```

#### Overall death ratio:-

```python
death_rate=(total_deaths/total_cases)*100
death_rate
```

#### Overall test conducted (Around 935310 no. of test were conducted before these data's being reported):-

```python
initial_test=935310
total_tests=initial_test+covid_df.new_tests.sum()
total_tests
```

#### Average ratio of positive results as per tests been conducted:-

```python
positive_rate=(total_cases/total_tests)
positive_rate
```

```python
print('{:.2f}%'.format(positive_rate*100))
```

### Working with dates

As it is a time series dataset, we can break it down to specific months or specific weeks.

```python
covid_df.date
```

As we can see here the datatype of dates mentioned as 'object', so we need to convert the datatype into 'datetime' format, 
now to do that we used 'pd.to_datetime' method.

```python
covid_df['date']=pd.to_datetime(covid_df.date)
```

```python
covid_df.date
```

To get different parts of date into separate columns :---

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```
